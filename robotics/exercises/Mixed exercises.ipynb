## Exercise 1

1. Consider the 2D robot in the figure, where you control the rotational speed of the left and right wheels $\dot{\varphi}_\text{left}$, $\dot{\varphi}_\text{right}$.  Wheel radius is 10 cm, and the wheel are 50 cm from each other.  Discuss the robot kinematics.  Define a suitable reference frame for the robot. 
2. Explain how you can compute the pose transformation that occurs when the robot moves for a time interval $\delta t$ with a given constant rotational speed for the left and right wheels $\dot{\varphi}_\text{left} \neq \dot{\varphi}_\text{right}$.
3. Consider a world reference frame $O$.  The robot is initially placed 1m to the right of the origin, pointing up.  Write the expression returning the final pose of the robot ${}^O T_{R}$ in the world frame of reference $O$, after the robot moves as follows:
  - for 2 seconds,  
  $\dot{\varphi}_\text{left} = \dot{\varphi}_\text{right} = 0.1 $ rad/s.
  - for 1 second,  
  $\dot{\varphi}_\text{left} = 0.1$ rad/s,   
  $\dot{\varphi}_\text{right} = 0.2 $ rad/s. 
4. Consider a point ${}^O P$ expressed in frame $O$. Explain how you can derive how much the robot should rotate in order to point straight to it.
